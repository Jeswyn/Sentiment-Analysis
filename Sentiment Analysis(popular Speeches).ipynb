In this particular project, we are going to work on the inaugural corpora from the nltk in Python. We will be looking at the following speeches of the Presidents of the United States of America:

<b>President John F. Kennedy in 1961</b>

<b>President Richard Nixon in 1973</b>

## Importing Necessary Libraries

```python
import numpy as np 
import pandas as pd 
import re 
import nltk 
import matplotlib.pyplot as plt
import string 
import matplotlib 
```

## Loading the documents from nltk corpus

```python

nltk.download('inaugural')
from nltk.corpus import inaugural
inaugural.fileids()
inaugural.raw('1941-Roosevelt.txt')
inaugural.raw('1961-Kennedy.txt')
inaugural.raw('1973-Nixon.txt') 
```

### Find the number of characters, words, and sentences for the mentioned documents. 

```python
for fileid in inaugural.fileids():
    num_chars = len(inaugural.raw(fileid))
    num_words = len(inaugural.words(fileid))
    num_sents = len(inaugural.sents(fileid))
    print(fileid)
    print('The no of characters is',num_chars, fileid )
    print('The no of words is',num_words,  fileid )
    print('The no of sentences are',num_sents , fileid )
```

### Remove all the stopwords from all three speeches. 

### 1941-Roosevelt Speech

```python
Roosevelt = nltk.corpus.inaugural.words('1941-Roosevelt.txt')


stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation) + ['--']

# Converting all the words to lower case

all_words = (w.lower() for w in Roosevelt)

# Only keeping the words which are not the 'stopwords'

all_words_clean = [word for word in all_words if word not in stopwords]

# Creating a frequency distribution of the lower case words which does not contain any stopwords

all_words_freq = nltk.FreqDist(all_words_clean)

# Extracting the most common 2000 words after the list of words have been converted to lowercase and the stopwords

word_features = [item[0] for item in all_words_freq.most_common(500)] 
```

```python
word_features 
```

```python
all_words_clean[:5] 
```

```python
all_words_freq 
```

#### Which word occurs the most number of times in hisinaugural address for Roosevelt ? Mention the top three words

```python
print('The word occurs the most number of times in hisinaugural address for Roosevelt are:')
word_features[0:3]
```

#### Plot the word cloud of each of the speeches of the variables - Roosevelt

```python
words_clean = ' '.join(word_features) 

from wordcloud import WordCloud , STOPWORDS
wordcloud = WordCloud(stopwords = STOPWORDS, width = 3000, height = 2500, 
                background_color ='white').generate(words_clean) 
  
# plot the WordCloud image                        
plt.figure(figsize = (20, 10), facecolor = 'k') 
plt.imshow(wordcloud) 
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad=0)
 

print('The world clous of roosevelt is')
plt.show()
```
